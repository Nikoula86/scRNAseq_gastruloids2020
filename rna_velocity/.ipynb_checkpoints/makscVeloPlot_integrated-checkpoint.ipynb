```python
def lighten_color(color, amount=0.5):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Examples:
    >> lighten_color('g', 0.3)
    >> lighten_color('#F034A3', 0.6)
    >> lighten_color((.3,.55,.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

```python
sample1 = scv.read(os.path.join(folder,"24h","possorted_genome_bam.loom"), cache=True)
sample2 = scv.read(os.path.join(folder,"48h","possorted_genome_bam.loom"), cache=True)
sample3 = scv.read(os.path.join(folder,"72h","possorted_genome_bam.loom"), cache=True)
```

```python
# sample1.var_names_make_unique()
# sample2.var_names_make_unique()
# sample3.var_names_make_unique()
```

```python
sample1.obs['x'] = [i[-17:-1]+'_24h_1' for i in sample1.obs.index]
sample2.obs['x'] = [i[-17:-1]+'_48h_1' for i in sample2.obs.index]
sample3.obs['x'] = [i[-17:-1]+'_72h_1' for i in sample3.obs.index]
```

```python
new_name = []
for i in range(len(cell_id)):
    if batch_ident.x.values[i]=='scRNA_gast24h':
        new_name.append(cell_id.x.values[i]+'_24h_1')
    elif batch_ident.x.values[i]=='scRNA_gast48h':
        new_name.append(cell_id.x.values[i]+'_48h_1')
    elif batch_ident.x.values[i]=='scRNA_gast72h':
        new_name.append(cell_id.x.values[i]+'_72h_1')
    
cell_id.x = new_name
```

```python
pd.DataFrame({'cid':cell_id.x.values,'batch':batch_ident.x.values})
```

```python
# split all cell ids into the three batches

cell_id1 = cell_id[batch_ident.x=='scRNA_gast24h']
cell_id2 = cell_id[batch_ident.x=='scRNA_gast48h']
cell_id3 = cell_id[batch_ident.x=='scRNA_gast72h']
```

```python
print(len(sample1),len(cell_id1))
print(len(sample2),len(cell_id2))
print(len(sample3),len(cell_id3))
print(len(sample1)+len(sample2)+len(sample3),len(cell_id1)+len(cell_id2)+len(cell_id3))
```

```python
# filter the loom to keep only cells annotated
sample1f = sample1[np.isin(sample1.obs.x, cell_id1.x)]
sample2f = sample2[np.isin(sample2.obs.x, cell_id2.x)]
sample3f = sample3[np.isin(sample3.obs.x, cell_id3.x)]
```

```python
print(len(sample1f))
print(len(sample2f))
print(len(sample3f))
print(len(sample1f)+len(sample2f)+len(sample3f))
```
